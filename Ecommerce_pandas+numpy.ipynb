```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df=pd.read_csv('ecommerce.csv')
```

```python
df.head(5)
```

```python
df.columns
```

```python
# 1- What are the total Order value for each country
df.groupby('Country').agg({
    'OrderValue':'sum'
}).reset_index()
```

```python
# 2- What each Major Category sold quantity during November 2011

df[df['Year-Month']=='2011-11   '].groupby(['Major Category','Year-Month']).agg({
    'Quantit':'sum'
}).reset_index()
```

```python
# 3- What's Total Order Value for each Minor Item in Kitchen

df[df['Major Category'].str.contains('  Kitchen ')].groupby('Major Category').agg({
    
    'OrderValue':'sum'
}).reset_index()

```

```python
# 4- What Month/s contributes to high demand in orders
df.groupby(['OrderValue','month_date']).agg({
    'Quantit':'max'
    
}).reset_index().sort_values(by='Quantit',ascending=False)
```

```python
# 5- What Month/s contributes to high OrderValue?
df.groupby('month_date').agg({
    'OrderValue':'max'
    
}).reset_index().sort_values(by='OrderValue',ascending=False)
```

```python
# 6- Why April and December suffer from low sales? 
temp=df.groupby(['month_date','Major Category','Minor Category','Country']).agg({
    'OrderValue':'sum'
}).reset_index()
temp['month_date'].isin([4,12])
temp[temp['month_date'].isin([4,12])]
```

```python
# 7- Why May and November have high sales? 
temp=df.groupby(['month_date','Major Category','Minor Category','Country']).agg({
    'OrderValue':'sum'
}).reset_index()
temp['month_date'].isin([5,11])
temp[temp['month_date'].isin([5,11])]
```

```python
df['conv_date']=pd.to_datetime(df['Year-Month'])
df['month_date']=df['conv_date'].dt.month
df['month_date']
```

```python
# 8- Most Major Items sold in April and December

most_major=df.groupby(['month_date','Major Category']).agg({
    'OrderValue':'max'    
}).reset_index()
most_major['month_date'].isin([4,12])
most_major[most_major['month_date'].isin([4,12])]
```

```python
# 9- Most Major Items sold in May and November
# same previous
most_major=df.groupby(['month_date','Major Category']).agg({
    'OrderValue':'max'    
}).reset_index()
most_major['month_date'].isin([5,11])
most_major[most_major['month_date'].isin([5,11])]
```
